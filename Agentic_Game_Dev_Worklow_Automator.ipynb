```python
# Installing dependencies
!pip install -q transformers sentencepiece wikipedia gradio

# Importing the required libraries
import wikipedia
import gradio as gr
from transformers import pipeline

# Loading the local LLM from Hugging Face
generator = pipeline("text2text-generation", model="google/flan-t5-base", tokenizer="google/flan-t5-base")

# Defining Wikipedia tool
def wiki_search(query):
    try:
        return wikipedia.summary(query, sentences=2)
    except:
        return "No Wikipedia summary found."

# Defining the full pipeline logic
def dev_agent(task):
    context = wiki_search(task)
    prompt = f"Use this info to help answer the request:\n\n{context}\n\nTask: {task}"
    output = generator(prompt, max_length=200, do_sample=False)[0]['generated_text']
    return output

# Loading the Gradio UI
gr.Interface(
    fn=dev_agent,
    inputs="text",
    outputs="text",
    title="🎮 Local Agentic Game Dev Assistant (Hugging Face)",
    description="Ask game dev-related questions (e.g., mythology, character ideas) – uses Wikipedia + local Hugging Face model"
).launch()
```
